<p style="font size:40px;">EXPLORATORY DATA ANALYSIS</p>

```python
ax=sns.countplot(x='Gender',data=df)
```

```python
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender',y='Amount',data=sales)
```

From above graph we can see that most of the buyers are females and even the purchasing power of female is grater than man

```python
df.columns
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
```

```python
ax=sns.countplot(x='Age Group',data=df,hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
#total amount vs age group
df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Age Group',y='Amount',data=sales_age)
```

From above graph we can see that most of the buyers are of age group between 26-35 years female

<h1>State</h1>

```python
df.columns
```

```python
#total number of orders from states
ax=sns.countplot(x='State',data=df,hue='Orders')
sns.set(rc={'figure.figsize':(30,5)})
```

```python
#total number of orders from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Orders',data=sales_state)
```

```python
#total amount/sales  from top 10 states
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(18,5)})
sns.barplot(x='State',y='Amount',data=sales_state)
```

From above graph we can say that most of the orders are from uttar pardesh ,maharashta and karnatka respectively

<h1>MARITAL STATUS </h1>

```python
df.columns
```

```python
ax=sns.countplot(x='Marital_Status',data=df)
sns.set(rc={'figure.figsize':(5,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(x='Marital_Status',y='Amount',data=sales,hue='Gender')
```

From above graph we can say that most of the buyers are married women and they have high purchasing power

<h1>OCCUPATION</h1>

```python
ax=sns.countplot(x='Occupation',data=df)
sns.set(rc={'figure.figsize':(30,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Occupation',y='Amount',data=sales)
```

From above graph we can say that most of the buyers are working in IT Sector and Healthcare sector

<h1>PRODUCT CATEGORY</h1>

```python
ax=sns.countplot(x='Product_Category',data=df)
sns.set(rc={'figure.figsize':(35,5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sales=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_Category',y='Amount',data=sales)
```

From above graph we can say that most of the sold products are from Food,Clothing and Electronic Gadget

```python
sales=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(x='Product_ID',y='Orders',data=sales)
```

```python
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar')
sns.set(rc={'figure.figsize':(25,5)})
```

<h1>Conclusion:</h1>

Married women age group 26-35 years from Up,Maharashtra and Karnatka woeking in It,Healthcare and Aviation are more likey to buy products from Food,clothing and Electronic category.
